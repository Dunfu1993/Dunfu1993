```python
def getGuessedWord(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters and underscores that represents
      what letters in secretWord have been guessed so far.
    '''
    # FILL IN YOUR CODE HERE...
    
    result = []
    for word in secretWord:
        if word in lettersGuessed:
            result.append(word)
        else:
            result.append("_ ")
    return "".join(result)
    
    
secretWord = 'apple'
lettersGuessed = ['e', 'i', 'k', 'p', 'r', 's']
print(getGuessedWord(secretWord, lettersGuessed))
```

```python


def getAvailableLetters(lettersGuessed):
    '''
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters that represents what letters have not
      yet been guessed.
    '''
    # FILL IN YOUR CODE HERE...
    import string
    result = []
    for letter in string.ascii_lowercase:
        if letter in lettersGuessed:
            continue
        else:
            result.append(letter)
    return "".join(result)

    
    
lettersGuessed = ['e', 'i', 'k', 'p', 'r', 's']
print(getAvailableLetters(lettersGuessed))
```

```python
def hangman(secretWord):
    '''
    secretWord: string, the secret word to guess.

    Starts up an interactive game of Hangman.

    * At the start of the game, let the user know how many 
      letters the secretWord contains.

    * Ask the user to supply one guess (i.e. letter) per round.

    * The user should receive feedback immediately after each guess 
      about whether their guess appears in the computers word.

    * After each round, you should also display to the user the 
      partially guessed word so far, as well as letters that the 
      user has not yet guessed.

    Follows the other limitations detailed in the problem write-up.
    '''
    # FILL IN YOUR CODE HERE...
    
    secretWord_length = len(secretWord)
    chances = 8
    lettersGuessed = []
    
    print("Welcome to the game Hangman!")
    print("I am thinking of a word that is " + str(secretWord_length) + " letters long.")
    print('---------------')
    
    while chances != 0:
        
        print('You have ' + str(chances) + ' left.')
        available_letters = getAvailableLetters(lettersGuessed)
        print('Available letters: ' + available_letters)
        current_letter = input('Please guess a letter: ')
        current_result = getGuessedWord(secretWord, lettersGuessed)
        
        if current_letter in lettersGuessed:
            print("Oops! You've already guessed that letter: " + current_result)
            continue
        else:
            lettersGuessed.append(current_letter)
            current_result = getGuessedWord(secretWord, lettersGuessed)

            if current_letter in list(secretWord):
                print('Good Guess: ' + current_result)
            else:
                print('Oops! That letter is not in my word: ' + current_result)
            
            print('---------------')
        
            chances -= 1
        
            if isWordGuessed(secretWord, lettersGuessed):
                print('Congratulations, you won!')
                break
        
        
        
    if chances == 0 and not isWordGuessed(secretWord, lettersGuessed):
        print('Sorry, you ran out of guesses. The word was else.')
    
        
        
hangman('dad')
```
